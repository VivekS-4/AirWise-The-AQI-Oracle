```python
def handle_outliers(df, column_name):
    '''Handles outliers in the specified column of the dataframe.'''

    # Calculate Q1, Q2 and IQR
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Define bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Handle outliers
    df[column_name] = df[column_name].apply(lambda x: lower_bound if x < lower_bound else x)
    df[column_name] = df[column_name].apply(lambda x: upper_bound if x > upper_bound else x)
    
    return df
```
