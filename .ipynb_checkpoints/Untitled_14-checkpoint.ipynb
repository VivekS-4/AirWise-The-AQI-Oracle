```python
import pandas as pd
import os

# Define the directory where your CSV files are located
csv_directory = 'C:/Users/HP/Desktop/AirWise/Toronto_West'

# Create an empty DataFrame to store the combined data
combined_data = pd.DataFrame()

# Loop through the CSV files in the directory
for filename in os.listdir(csv_directory):
    if filename.endswith(".csv"):
        file_path = os.path.join(csv_directory, filename)
        
        # Read each CSV file into a DataFrame
        data = pd.read_csv(file_path)
        
        # Append the data to the combined_data DataFrame
        combined_data = combined_data.append(data, ignore_index=True)

# Specify the path for the new combined CSV file
combined_csv_file = 'combined_data.csv'

# Save the combined data to a new CSV file
combined_data.to_csv(combined_csv_file, index=False)

print(f"Combined data saved to {combined_csv_file}")
```

```python
import pandas as pd
import os

# Path to the directory containing the CSV files
input_directory = "C:/Users/HP/Desktop/AirWise/Toronto_West"

# List of CSV files to read and append
csv_files = [file for file in os.listdir(input_directory) if file.endswith(".csv")]

# Path to the output CSV file
output_file = "output.csv"
```

```python
# Create an empty DataFrame
result_df = pd.DataFrame()
```

```python
for csv_file in csv_files:
    # Construct the full path to the CSV file
    file_path = os.path.join(input_directory, csv_file)
    
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    
    # Append the DataFrame to the result DataFrame
    result_df = result_df.append(df, ignore_index=False)
```

```python
# Save the result DataFrame to the output CSV file
result_df.to_csv(output_file, index=False)
```

```python

# Specify the name of your zip file containing CSVs
zip_file = 'C:\\Users\\nabil\\Downloads\\Toronto_West.zip'


# Create an empty list to store the DataFrames
dataframes = []

# Open the zip file
with zipfile.ZipFile(zip_file, 'r') as archive:
    # Loop through the files in the zip archive
    for file_name in archive.namelist():
        # Check if the file is a CSV file (you can add more specific checks if needed)
        if file_name.endswith('.csv'):
            # Read the CSV file from the zip archive into a DataFrame
            with archive.open(file_name) as file:
                data = pd.read_csv(file)
                # Check if the 'category' column exists, and if it does, drop it
                if 'category' in data.columns:
                    data = data.drop(columns=['category'])
                # Transpose the data
                data = data.T
                data.reset_index(drop=True, inplace=True)
                dataframes.append(data)

# Concatenate the list of DataFrames into a single DataFrame
merged_data = pd.concat(dataframes, ignore_index=True, axis=1)

```

```python
# Save the merged and transposed data to a new CSV file
merged_data.to_csv('C:\Users\nabil\OneDrive\Desktop\toronto_west\torontowest.xlsx', index=False)
```
